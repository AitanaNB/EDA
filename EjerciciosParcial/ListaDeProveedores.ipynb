```java
public class Nodo
{
    String nombreDeProveedor;
    int cantidad;
    Nodo sigProv;
    Nodo superior;

    public Nodo(String pNombre, int cantidad)
    {
        this.nombreDeProveedor=pNombre;
        this.cantidad=cantidad;
    }
}

public class ListaDeProveedores
{
    Nodo first;

    public void eliminarProveedor(String proveedor)
    {
        //pre: el proveedor está en la estructura, no es el único prov del nivel
        //ningún proveedor de un nivel inferiorr (externo) a punta a ese prov
        //post: se ha eliminado el proveedor

        Boolean borrado=false;
        Nodo nivelActual= this.first;

        while(!borrado && nivelActual.sigProv!=nivelActual)
        {
            Nodo actual=nivelActual.sigProv;
            Nodo anterior=actual;
            Boolean enc= false;
            do
            {
                if(actual.nombreDeProveedor.equals(proveedor))
                {
                    enc=true;
                }
                else
                {
                    anterior=actual;
                    actual=actual.sigProv;
                }
            }while(actual!=nivelActual && !enc);

            if(!enc) //si hay que pasar de nivel
            {
                nivelActual=nivelActual.superior;
            }
            else //si hay que borrar
            {
                borrado=true;
                if(actual==first) //actual es el primero
                {
                    anterior.sigProv=actual.sigProv;
                    first=actual.sigProv;
                }
                else //es un nodo normal
                {
                    anterior.sigProv=actual.sigProv;
                }
            }
        }


    }
    
}
```
