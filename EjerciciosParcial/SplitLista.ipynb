Dividir Listas par e impar

```java
public class Node 
{
    int data;
    Node next;

    public Node (int data)
    {
        this.data=data;
    }
}

public class DoubleNode 
{
    int data;
    DoubleNode next;
    DoubleNode prev;

    public DoubleNode(int data) {
        this.data = data;
    }
}

public class ListaX
{
    Node first1; //lista simple
    DoubleNode first2; //lista doblemente enlazada

    public void split()
    {
        //Pre: DDlist (first2) es null
        //post: se han quitado los elementos impares de lista1, y se han puesto en l2

        if (first1 == null){}
        else
        {
            Node actual = first1;
            Node anterior = null;

            DoubleNode ultimo =null;
            

            while(actual!=null)
            {
                if(actual.data %2 !=0) //si el dato es impar
                {
                    //1- borrar el dato de lista1 y guardar valor
                    DoubleNode nuevoImpar = new DoubleNode(actual.data); //guardar el valor impar
                    if(actual==first1)//si el impar es el primero
                    {
                        first1=first1.next;
                    }
                    else
                    {
                        anterior.next=actual.next;
                    }

                    //2- añadir el dato a lista2
                    if(first2==null)
                    {
                        first2=nuevoImpar;
                    }
                    else
                    {
                        ultimo.next=nuevoImpar;
                        nuevoImpar.prev=ultimo;
                    }
                    ultimo=nuevoImpar;
                    actual=actual.next;
                }
                else //si es par
                {
                    anterior=actual;
                    actual=actual.next;
                }
            }
        }
    }

    public void imprimir1()
    {
        Node actual= first1;
         System.out.println("Lista 1: ");
        while (actual!=null)
        {
            System.out.println(actual.data);
            actual=actual.next;
        }
    }
    public void imprimir2() {
        DoubleNode actual = first2;
        System.out.println("Lista 2: ");
        if (actual == null) {
            System.out.println("La lista está vacía.");
            return;
        }
    
        // Imprime el primer nodo
        System.out.println(actual.data);
        actual = actual.next;
    
        // Recorre la lista hasta volver al primer nodo o llegar a un final (en no circulares)
        while (actual != null && actual != first2) {
            System.out.println(actual.data);
            actual = actual.next;
        }
    }
    
    
}
```
