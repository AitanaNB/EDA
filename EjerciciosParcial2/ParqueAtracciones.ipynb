Parque de atracciones

```java
public class Accion
{
    char tipo; //(i llegar, s entrar, e usar)
    String dni; //tipo i
    int ventana; //i o s
    int atraccion; //e
    Queue<Integer> atracciones;  //i
}

public class Cliente
{
    String dni;
    Queue<Integer> atracciones;

    public Cliente(String dni, Queue<Integer> atracciones)
    {
        dni=dni;
        atracciones = atracciones;
    }
}
```

```java
public ArrayList<String> simularParque(ArrayList<Accion> listaAcciones)
{
    //Pre: cada cliente como minimo 1 atraccion en lista
    //Post: devuelve lista de clientes que han subido a todas sus atracciones

    //10 ventanas, 20 atracciones
    ArrayList <String> completados = new ArrayList<>();

    //inicializar ventanas y atracciones
    Queue<Cliente> [] ventanas= new LinkedList [10];
    for(int i=0; i<10; i++)
    {
        ventanas[i]=new LinkedList<Cliente>();
    }

    Queue<Cliente> [] atracciones= new LinkedList [20];
    for(int j=0; j<20; j++)
    {
        atracciones[j]=new LinkedList<Cliente>();
    }

    //ir acci贸n por acci贸n
    for(Accion accion: listaAcciones) 
    {
        //Llegar: entra a la ventana
        if(accion.tipo == 'I')
        {
            Cliente cliente = new Cliente(accion.dni, accion.atracciones);
            ventanas[accion.ventana].add(cliente);
        }
        //Entrar: sale de la ventana, entra a atraccion
         else if(accion.tipo == 'S')
        {
            Cliente cliente = ventanas[accion.ventana].poll();
            int atraccion=cliente.atracciones.peek();
            atracciones[atraccion].add(cliente);
        }
        //Usar atracci贸n: sale de atraccion
        else if(accion.tipo == 'E')
        {
            Cliente cliente = atracciones[accion.atraccion].poll();
            //quitar atraccion de la lista de atracciones del cliente-> ya se ha subido 
            cliente.atracciones.poll();
            if(cliente.atracciones.isEmpty()) //si esta completado
            {
                completados.add(cliente.dni);
            }
            else //cuando baja pasa a la pr贸xima atraccion directamente
            {
                int siguienteAtr= cliente.atracciones.peek();
                atracciones[siguienteAtr].offer(cliente);
            }
        }
    }

    return completados;
}
```
