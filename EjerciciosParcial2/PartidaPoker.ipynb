```java
public class Pago
{
    int pagador;
    int cobrador;
    int cantidad;

    public Pago(int p, int cob, int cant)
    {
        this.pagador=p;
        this.cobrador=cob;
        this.cantidad=cant;
    }
}
public class Juego {
	   
    public int jugar(int nJugadores, int nBilletesInicial, ArrayList<Pago> pagos) 
    {
    // Precondición:  2 <=  nJugadores <= 20
    //                nBilletesInicial >= 1
    //                Todos los pagos de la lista “pagos” son correctos,
    //                 es decir, un jugador siempre tiene los billetes 
    //                 necesarios para realizar sus pagos 
    // Postcondición: el resultado es el número de billetes verdaderos 
    //                 del jugador 0 al acabar la partida
        /*
         * incializar jugadores
         * jugar
         * contar billetes jugador
         */
    	
        // cada jugador tiene stack de billetes verdadero o falso
        @SuppressWarnings("unchecked")
        Stack<Boolean>[] jugadores = new Stack[nJugadores];
       
        //hay que crear las pilas del array, sino serían null
        for(int i=0; i<nJugadores; i++)
        {
            jugadores[i]=new Stack <Boolean>();
        }
        
        //meter billetes iniciales por jugador
        for(int i=0; i<nJugadores; i++)
        {
            for(int j=0; j<nBilletesInicial; j++)
            { // True si no es el tahúr, False si es el tahúr (o)
                boolean billete= (i!=0);
                jugadores[i].push(billete);
            }
        }

        //leer los pagos
        for(Pago pagoAct: pagos)
        {   
            //simular todos los pagos
            for(int k=1; k<=pagoAct.cantidad; k++)
            {
                //sacar el último billete del pagador
                boolean billete=jugadores[pagoAct.pagador].pop();
                //añadirlo a la lista del cobrador
                jugadores[pagoAct.cobrador].push(billete);
            }
        }

        int gananciasTahur=0;

        //contar los billetes del tahúr(0)
        while(!jugadores[0].isEmpty())
        {
            boolean billete=jugadores[0].pop();
            if(billete)
            {
                gananciasTahur++;
            }
        }
        
        return gananciasTahur;
    }
    public ArrayList<Pago> crearListaPagos(){
        /*    	Inicio: 
        0: 50F
        1: 50T
        2: 50T
    
        0 1 10 -->
        0: 40F
        1: 50T + 10 F
        2: 50T
    
        1 2 15 -->
        0: 40F
        1: 45T
        2: 50T + 10F + 5T
    
        2 0 2 -->
        0: 40F + 2T
        1: 45T
        2: 50T + 10F + 3T
    
        1 2 20 -->
        0: 40F + 2T
        1: 25T
        2: 50T + 10F + 3T + 20T
    
        1 0 2 -->
        0: 40F + 2T + 2T
        1: 23T
        2: 50T + 10F + 3T + 20T
    
        2 0 5 -->
        0: 40F + 2T + 2T + 5T
        1: 23T
        2: 50T + 10F + 3T + 15T
        */
        ArrayList<Pago> pagos = new ArrayList<Pago>();
        pagos.add(new Pago(0, 1, 10));
        pagos.add(new Pago(1, 2, 15));
        pagos.add(new Pago(2, 0, 2));
        pagos.add(new Pago(1, 2, 20));
        pagos.add(new Pago(1, 0, 2));
        pagos.add(new Pago(2, 0, 5));
        
        return pagos;
    }
}
```

```java

Juego juego = new Juego();
ArrayList<Pago> pagos = juego.crearListaPagos();

System.out.println("Prueba: pagos (0 1 10), (1 2 15), (2 0 2), (1 2 20), (1 0 2), (2 0 5)");
System.out.println("El resultado debería ser 9 y es: " + juego.jugar(3, 50, pagos));
```
